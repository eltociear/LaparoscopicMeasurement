```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="whitegrid")
sns.set_context("paper")

# Create the plot
plt.figure(figsize=(6, 9))  # Adjust the size of the plot
sns.violinplot(data=df, 
x="Mode", y="Value", hue="Selection",
               split=True, inner="quart", linewidth=1.3, palette="pastel")  # I added the palette for a finer color distinction

# Adjust styles and labels
sns.despine(left=True)
plt.title('Comparison of Measurement Functionalities and Point Selection Types', fontsize=16)  # Add a title for the entire plot
plt.xlabel('Functionality', fontsize=14)  # Add a title for the x-axis
plt.ylabel('Error in mm', fontsize=14)  # Add a title for the y-axis
plt.legend(title='Type')  # Add a title to the legend

# Set y-axis limits and grid resolution
plt.ylim(-10, 10)
plt.yticks(np.arange(-10,11, 2))  # This will set y-ticks at intervals of 0.5 from -10 to 10

plt.show()  # Display the plot

# Save the figure
plt.savefig('measurements.pdf')
```
